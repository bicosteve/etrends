```python
# Create x which contains the features to use for training your mode
# This defines the features to be used in training the model
X = df[['MonsoonIntensity', 'TopographyDrainage', 'RiverManagement',
       'Deforestation', 'Urbanization', 'ClimateChange', 'DamsQuality',
       'Siltation', 'AgriculturalPractices', 'Encroachments', 'DrainageSystems',
       'CoastalVulnerability', 'Landslides', 'Watersheds',
       'DeterioratingInfrastructure', 'WetlandLoss']]
```

```python
# Y is the target variable
y = df['FloodProbability']
```

```python
# Train test split for training and testing the data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Creating and training the model
from sklearn.linear_model import LinearRegression
```

```python
# Instantiate the model
model = LinearRegression()
```

```python
# Train the model using X_train and y_train from dataset
model.fit(X_train, y_train)
```

```python
# Evaluating the model
# 1. Print the intercept
print('Model intercept ---->',model.intercept_)
```

```python
# 2. Model Coefficient
model.coef_
```

```python
X_train.columns
```

```python
coef_df = pd.DataFrame(model.coef_, X.columns, columns=['Coeff'])
```

```python
#print coefficient for each of the features
coef_df
```

```python
# Make Predictions from test set 
# Use features which the model has never seen before
predictions = model.predict(X_test)
```

```python
predictions
```

```python
# Y_test contains the correct predictions of flooding
y_test
```

```python
# Plot a scatter to view how far off or close the predictions are from the reality of flooding
# If there is nearly a perfect straight line then it means the model did a good job in making predictions
plt.scatter(y_test, predictions, color='blue')
plt.title('Linear Regression for flood prediction')
plt.xlabel('predictions')
plt.ylabel('actual values')
plt.show()
```

```python
#Histogram for distribution of the residuals
sns.distplot((y_test-predictions))
plt.show()
```

```python
# Since the residuals are normally distributed, it means the linear regression algorithm was the best choice for 
# flood predictions model.
# 
```

```python
# Evaluation Metrics
# Common Regression Metrics evaluation 
# 1. Mean Absolute Error = MAE --> is the mean of absolute values of error
# 2. Mean Squared Error = MSE --> the mean of squared errors 
# 3. Root Mean Error = RMSE --> is the square root of the mean of the squared errors

from sklearn import metrics

print('MAE ---> ', metrics.mean_absolute_error(y_test, predictions))
print('MSE ---> ', metrics.mean_squared_error(y_test, predictions))
print('RMSE ---> ', np.sqrt(metrics.mean_absolute_error(y_test, predictions)))
```
