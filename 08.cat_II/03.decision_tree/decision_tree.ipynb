```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
data_frame = pd.read_csv('../dataset/cat_2_dataset.csv')
```

```python
data_frame.columns
```

```python
X = data_frame[['Study Hours', 'Sleep Hours', 'Attendance','Assignment Completed', 'Internet Usage (hrs)']]
```

```python
# Getting the values of y
# You can use labelEncoder to transform the data into numeric or just make it have 1 and 0 according to
# pass status
y = []
for score in data_frame['Exam Scores']:
    if score < 50:
        y.append(1)
    else:
        y.append(0)
print(y)
```

```python
model = DecisionTreeClassifier()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=10)
```

```python
# Training the model
model.fit(X_train, y_train)
```

```python
# Making predictions with x_test
prediction = model.predict(X_test)
```

```python
# Evaluate Prediction
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, prediction)
report = classification_report(y_test, prediction)
confusion_mtrx = confusion_matrix(y_test, prediction)

print(f"Model accuracy {accuracy}")
print(f"Model classification report {report}")
print(f"Model confusion matrix {confusion_mtrx}")
```

```python
# Find prediction for provided data
data = [5,6,85,4,3]
data = np.array(data)

new_prediction = model.predict(data.reshape(1,-1))
print("New prediction is",new_prediction)
```
